# date_parser

```python
from datetime import datetime

def parser(x,y):
    dt = x +' '+ y.split('.')[0]
    return datetime.strptime(dt, '%d/%m/%Y %H')
```

```python
df = pd.read_csv('./data/AirQualityUCI/AirQualityUCI.csv',delimiter=';',parse_dates=[['Date','Time']], date_parser=parser)
df
```
