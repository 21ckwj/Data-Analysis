```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
# 데이터셋
from darts.datasets import SunspotsDataset
df_sun = SunspotsDataset().load().pd_dataframe()
df_sun = df_sun.loc['1900-01':'1933-12']
df_sun
```

```python
# 시각화
df_sun.plot(figsize=(10,5),title='SunSpots')
```

```python
df_sun['year'] = df_sun.index.year
df_sun['month'] = df_sun.index.month
df_sun.head()
```

# 연도별 흑점개수

```python
plt.figure(figsize=(13,7))
plt.xticks(rotation=30)
ax = sns.barplot(data=df_sun, x='year', y='Sunspots')
```

```python
plt.figure(figsize=(13,7))
plt.xticks(rotation=30)
ax = sns.barplot(data=df_sun.loc['1901-01':'1913-12'], x='year', y='Sunspots', hue='month')
```

# 월별 흑점수

```python
plt.figure(figsize=(13,7))
plt.xticks(rotation=30)
ax = sns.barplot(data=df_sun.loc['1901-01':'1913-12'], x='month', y='Sunspots')
```
