# Trend

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from darts.datasets import AusBeerDataset

sns.set_theme(context='talk', color_codes=True)
```

```python
df_beer = AusBeerDataset().load().pd_dataframe()
df_beer.head()
```

```python
fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
df_beer.plot(ax=ax)
ax.set_title('Beer sales')
```

## trend 추출 by resample

```python
df_beer_y = df_beer.resample('Y').mean()
df_beer_y.head()
```

```python
fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
ax.plot(df_beer.index,df_beer['Y'],label='3M')
ax.plot(df_beer_y.index, df_beer_y['Y'],label='Yearly')
ax.set_title('Beer sales')
ax.legend()
```

## trend 추출 by rolling

```python
window_size = 4
df_beer_y1 = df_beer.rolling(window=window_size).mean()
df_beer_y1.head()
```

```python
fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
ax.plot(df_beer.index,df_beer['Y'],label= '3M')
ax.plot(df_beer_y1.index, df_beer_y1['Y'],label= f'window={window_size}')
ax.set_title('Beer sales')
ax.legend()
```

## trend 추출 by rolling(center=True)

```python
window_size = 4
df_beer_y2 = df_beer.rolling(window=window_size,center=True).mean()
df_beer_y2.head()

fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
ax.plot(df_beer.index,df_beer['Y'],label= '3M')
ax.plot(df_beer_y2.index, df_beer_y2['Y'],label= f'window={window_size}')
ax.set_title('Beer sales')
ax.legend()
```

# trend 추출 by lowess

https://www.youtube.com/watch?v=Vf7oJ6z2LCc

* 데이터 손실이 없는이유?
-> 평균이 아닌 window_size 박스 안에서 weighted least squares을 통해 추세선을 그리기 때문
-> 처음 몇개의 데이터는 같은 window_size 박스를 사용

* outlier(이상치)에 강한 이유?
-> 원래 기준점과 weighted least squares를 사용하여 구한 점의 차이를 조정해 주기 때문
-> 원래 기준점과 새로 구한 점의 차이가 작을수록 더 높은 weight를 가지는 새로운 회귀 곡선을 그려줌


## 실습

```python
from statsmodels.nonparametric.smoothers_lowess import lowess
y = df_beer['Y']
x = np.arange(0,len(y))

# https://www.statsmodels.org/dev/endog_exog.html
# endogenous: caused by factors within the system 내생변수
# exogenous: caused by factors outside the system 외생변수

res = lowess(endog=y,
            exog=x,
            frac=0.1,
            it=3)
res 
```

```python
df_beer['lowess'] = res[:,1]
df_beer
```

```python
fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
df_beer.plot(ax=ax)
ax.set_title('Beer sales')
```

```python
 from statsmodels.nonparametric.smoothers_lowess import lowess
y = df_beer['Y']
x = np.arange(0,len(y))

# https://www.statsmodels.org/dev/endog_exog.html
# endogenous: caused by factors within the system 내생변수
# exogenous: caused by factors outside the system 외생변수

# https://www.statsmodels.org/dev/generated/statsmodels.nonparametric.smoothers_lowess.lowess.html
# Between 0 and 1. The fraction of the data used when estimating each y-value.

frac = 0.03 # 0,1 사이, 각 y값을 추정하는데 사용되는 data의 부분
            # 여기서의 y값은 회귀 직선이므로 frac 작을수록 
res = lowess(endog=y,
            exog=x,
            frac= frac,
            it=3)

df_beer['lowess'] = res[:,1]
fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
df_beer.plot(ax=ax)
ax.set_title('Beer sales')
```
