# Trend

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from darts.datasets import AusBeerDataset

sns.set_theme(context='talk', color_codes=True)
```

```python
df_beer = AusBeerDataset().load().pd_dataframe()
df_beer.head()
```

```python
fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
df_beer.plot(ax=ax)
ax.set_title('Beer sales')
```

## trend  by resample

```python
df_beer_y = df_beer.resample('Y').mean()
df_beer_y.head()
```

```python
fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
ax.plot(df_beer.index,df_beer['Y'],label='3M')
ax.plot(df_beer_y.index, df_beer_y['Y'],label='Yearly')
ax.set_title('Beer sales')
ax.legend()
```

## trend  by rolling

```python
window_size = 4
df_beer_y1 = df_beer.rolling(window=window_size).mean()
df_beer_y1.head()
```

```python
fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
ax.plot(df_beer.index,df_beer['Y'],label= '3M')
ax.plot(df_beer_y1.index, df_beer_y1['Y'],label= f'window={window_size}')
ax.set_title('Beer sales')
ax.legend()
```

## trend  by rolling(center=True)

```python
window_size = 4
df_beer_y2 = df_beer.rolling(window=window_size,center=True).mean()
df_beer_y2.head()

fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
ax.plot(df_beer.index,df_beer['Y'],label= '3M')
ax.plot(df_beer_y2.index, df_beer_y2['Y'],label= f'window={window_size}')
ax.set_title('Beer sales')
ax.legend()
```

```python
from statsmodels.nonparametric.smoothers_lowess import lowess
y = df_beer['Y']
x = np.arange(0,len(y))

# https://www.statsmodels.org/dev/endog_exog.html
# endogenous: caused by factors within the system 
# exogenous: caused by factors outside the system 

res = lowess(endog=y,
            exog=x,
            frac=0.1,
            it=3)
res 
```

```python
df_beer['lowess'] = res[:,1]
df_beer
```

```python
fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
df_beer.plot(ax=ax)
ax.set_title('Beer sales')
```

```python
 from statsmodels.nonparametric.smoothers_lowess import lowess
y = df_beer['Y']
x = np.arange(0,len(y))

# https://www.statsmodels.org/dev/endog_exog.html
# endogenous: caused by factors within the system 
# exogenous: caused by factors outside the system 

# https://www.statsmodels.org/dev/generated/statsmodels.nonparametric.smoothers_lowess.lowess.html
# Between 0 and 1. The fraction of the data used when estimating each y-value.

frac = 0.03 # 0,1 ,  y   data 
            #  y   frac  
res = lowess(endog=y,
            exog=x,
            frac= frac,
            it=3)

df_beer['lowess'] = res[:,1]
fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
df_beer.plot(ax=ax)
ax.set_title('Beer sales')
```
