```python
import scipy as sp
import matplotlib.pyplot as plt

def plot_binom_distribution(N, p):
    binom = sp.stats.binom(N, p)
    # 확률변수
    xx = np.arange(N + 1)
    plt.figure(figsize=(8,5))
    plt.bar(xx, binom.pmf(xx), align="center")
    plt.xlabel("표본값")
    plt.ylabel("$P(x)$")
    plt.title("이항분포의 확률질량함수")
    plt.show()
```

```python
N = 50
p = 0.3
plot_binom_distribution(N, p)
```
