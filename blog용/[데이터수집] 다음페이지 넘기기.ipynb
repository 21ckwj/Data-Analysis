# 쉬핑가제트

```python
url1 = 'https://www.ksg.co.kr/shippingGraph/hrci_graph.jsp'
driver.get(url1)
html = driver.page_source
df_all = pd.read_html(html)[0]
print(df_all)

url2 = 'https://www.ksg.co.kr/shippingGraph/hrci_graph.jsp?pageNum=2&schEDate=2022-12-08'
driver.get(url2)
time.sleep(2)

last_page = int(driver.find_element(By.CSS_SELECTOR,'#container > div.contents > div.fl_cont > div.sub_cont > div.f11.gray > strong:nth-child(2)').text)

page_bar = driver.find_elements(By.CSS_SELECTOR, '#container > div.contents > div.fl_cont > div.sub_cont > div.paging> span>a')
for page_num in range(2,last_page+1):
    print(page_num)
    
    for i in range(2,len(page_bar)-2): # 처음페이지,이전페이지 .. 제외한 인덱스 : 2~11
        
        if int(page_bar[i].text) == page_num:
            page_bar[i].click()
            time.sleep(2)
            html = driver.page_source
            df = pd.read_html(html)[0]
            df_all = pd.concat([df_all,df])
            print(df_all[-5:])
            
            page_bar = driver.find_elements(By.CSS_SELECTOR, '#container > div.contents > div.fl_cont > div.sub_cont > div.paging> span>a')
            print('클릭한 페이지:',page_bar[i].text)
            
            if (int(page_bar[i].text) % 10 ==0):  # 다음10페이지로 버튼 클릭  
                driver.find_element(By.CSS_SELECTOR,'#container > div.contents > div.fl_cont > div.sub_cont > div.paging > span.last > a').click()
                page_bar = driver.find_elements(By.CSS_SELECTOR, '#container > div.contents > div.fl_cont > div.sub_cont > div.paging> span>a')
                
            if (int(page_bar[i].text) == last_page):
                break
```
