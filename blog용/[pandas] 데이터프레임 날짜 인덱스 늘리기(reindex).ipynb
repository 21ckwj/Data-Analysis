```python
df = sns.load_dataset("flights")
df
```

```python
# 문자열 -> 숫자형
df['month'] = pd.factorize(df['month'])[0] +1

# 숫자 -> 문자열
df['year'] = df['year'].astype('str')
df['month'] = df['month'].astype('str')

# 문자열 -> datetime 형식 인덱스 생성 
df['date'] = df['year'] + '-' + df['month']
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date')

df_a = df['passengers'].loc["1950-01":"1950-08"].to_frame()
df_b = df['passengers'].loc["1950-01":"1950-06"].to_frame()
```

```python
df_a
```

```python
# 새로운 날짜 범위 지정
new_date_range = df_a.index
# reindex 함수 사용하여 늘려주기
df_b.reindex(new_date_range)
```
